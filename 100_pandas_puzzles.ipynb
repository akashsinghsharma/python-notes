# DataFrame basics

Fundamental routines for selecting, sorting, adding and aggregating data in DataFrames

```python
import numpy as np
```

```python
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
# Make labels out of this data and index should be from labels.

df1 = pd.DataFrame(data = data, index = labels)
df1
```

```python
# Create a summary of this data frame.

# Method 1:

df1.describe()
```

```python
#Method 2:

df1.info()
```
